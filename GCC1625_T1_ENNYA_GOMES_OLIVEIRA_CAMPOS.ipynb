###(i) Esboce um gráfico (histograma) da distribuição da população com relação à característica quantidade de páginas. Dica: observe que esta população é modelada por uma distribuição discreta uniforme. Veja detalhes nesta entrada da Wikipédia1.


```python
import matplotlib.pyplot as plt
import numpy as np

# população: inteiros de 150 a 300
pop = np.arange(150, 301)

# probabilidades iguais
probs = np.ones_like(pop) / len(pop)

plt.figure(figsize=(10, 4))
plt.bar(pop, probs, width=1, color='#ADD8E6', edgecolor='black')
plt.title("Distribuição Uniforme Discreta - Número de Páginas dos Livros")
plt.xlabel("Número de Páginas", fontsize=12)
plt.ylabel("Probabilidade", fontsize=12)
# Ticks do eixo X a cada 10 páginas
plt.xticks(np.arange(150, 301, 10))

# Ticks do eixo Y de 0 até 0.007, de 0.001 em 0.001
plt.yticks(np.arange(0, 0.0075, 0.001))
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

Explicação:

A população é formada por livros com número de páginas entre 150 a 300. Todos os números dentro desse intervalo têm a mesma chance de acontecer, o que caracteriza uma distribuição uniforme discreta.

No gráfico, cada barra mostra a quantidade de páginas de cada livro dentro do intervalo de 150 a 300. Como é uma distribuição uniforme, todas as barras têm a mesma altura, o que significa que todos os valores têm a mesma probabilidade de acontecer, que é 1/151 ≈ 0,0066.


###(ii) Encontre a média (µ), a variância (σ2) e o desvio padrão (σ) da distribuição da população. Dica: use as fórmulas da distribuição uniforme discreta disponíveis no link acima.


```python
# Limites inferior e superior da distribuição uniforme
a = 150
b = 300

# Fórmulas da distribuição uniforme discreta
media = (a + b) / 2
variancia = ((b - a + 1)**2 - 1) / 12
desvio_padrao = np.sqrt(variancia)

print(f"Média (μ): {media}")
print(f"Variância (σ²): {variancia}")
print(f"Desvio padrão (σ): {desvio_padrao}")
```

Explicação:

Primeiro, foram calculados os parâmetros que representam o número de páginas dos livros, 150 e 300.

Por ser uma distribuição uniforme discreta, foram aplicadas as fórmulas para os parâmetros populacionais:

Média (μ):
μ = (a+b)/2

Que representa o valor central da distribuição.
Onde a = 150 e b = 300, resultando em 225.

Variância (σ²):
σ² =  ((b − a + 1)² - 1)/ 2

Que mede o grau de dispersão dos valores em torno da média. Resultando em 1900.

Desvio padrão:
É a raiz quadrada da variância, e indica o quanto os valores se afastam da média. Resultando em 43.59.

###(iii) Considere que o plano amostral utilizado é uma amostragem aleatória simples com reposição (simple random sample with replacement). Considere tomar amostras de tamanho 2 dessa população e calcular a média de cada amostra. Para cada amostra, você terá uma estatística pontual (a média amostral x). Ao gerar todas as possíveis amostras aleatórias de tamanho 2, a distribuição de todas essas médias constitui a distribuição amostral da média (sampling distribution of the sample mean) para n = 2.
###• Primeiro, implemente uma função que gere todas as possíveis amostras de tamanho 2 com reposição, e compute a média de cada uma. Por exemplo, para as amostras (200, 200) e (300, 250), as médias seriam 200 e 275, respectivamente. Note que diferentes amostras podem produzir a mesma média (como, por exemplo, (250, 300), (275, 275), (280, 270), etc.).
###• Em seguida, usando os valores das médias obtidas, esboce um histograma da distribuição amostral da média.



```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import product

# População
pop = np.arange(150, 301)  # 150 até 300 inclusive

# Gera todas as amostras possíveis de tamanho 2 com reposição
amostras = list(product(pop, repeat=2))  # produto cartesiano com repetição

# Calcula a média de cada amostra
medias_amostrais = [np.mean(amostra) for amostra in amostras]

print(f"Quantidade de amostraas: {len(amostras)}")
print(f"Quantidade de médias amostrais: {len(medias_amostrais)}")

# Histograma das médias amostrais
plt.figure(figsize=(12, 5))
plt.hist(medias_amostrais, bins=50, edgecolor='black', color='lightgray')
plt.title("Distribuição amostral da média (n = 2)")
plt.xlabel("Média da amostra")
plt.ylabel("Frequência")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

Explicação:

Foram geradas todas as 22.801 amostras possíveis de tamanho 2 com reposição a partir da população de 150 a 300. Para cada amostra, foi calculada a média, resultando também em 22.801 médias amostrais, com 151 valores distintos.

O código utiliza a função product(pop, repeat=2) da biblioteca itertools para gerar todas as combinações possíveis de tamanho 2. Cada amostra gerada é uma tupla com dois números representando a quantidade de páginas dos livros. Em seguida, foi calculada a média de cada amostra usando np.mean(amostra) e armazenada as médias em uma lista.

Com todas as médias amostrais calculadas, foi gerado um histograma para visualizar a distribuição dessas médias. O histograma permite observar como as médias das amostras de tamanho 2 estão distribuídas. Como a amostragem é aleatória com reposição, a distribuição amostral da média tende a ser centralizada, refletindo a tendência das amostras geradas. O histograma tem um formato triangular, com maior frequência de médias em torno de 225, que é a média da população.

# (2) Problema dos tanques alemães
###(i) Apresente uma estimativa pontual de N com base na amostra coletada, usando o seguinte estimador clássico:
##Nˆ = M + M / k − 1
###onde M é o maior número observado na amostra, e k é o tamanho da amostra.


```python
amostra = [17, 32, 45, 59, 88]
k = len(amostra)
M = max(amostra)

# Estimativa pontual usando o estimador clássico
N_hat = M + M / k - 1

print(f"Maior número da amostra (M): {M}")
print(f"Estimativa de N (N̂): {N_hat:.2f}")
```

Explicação:

O código calcula uma estimativa do número total de tanques (N) com base em uma amostra de 5 tanques capturados. A fórmula usada leva em conta o maior número de série da amostra (88) e o número de tanques capturados (5). A estimativa de N foi calculada como 104,6, o que significa que, com base nessa amostra, estima-se que o inimigo tenha produzido cerca de 104 tanques.

###(ii) Implemente (ou descreva) uma simulação computacional que reproduza o processo de amostragem aleatória de k = 5 tanques a partir de uma população com N = 100 tanques numerados sequencialmente. Repita esse processo muitas vezes (por exemplo, 10.000 vezes) e registre as estimativas Nˆ em cada simulação.


```python
import numpy as np

N_real = 100   # número real de tanques
k = 5          # tamanho da amostra
n_sim = 10000  # número de simulações

estimativas = []

for _ in range(n_sim):
    amostra = np.random.choice(np.arange(1, N_real + 1), size=k, replace=False)
    M = max(amostra)
    N_hat = M + M / k - 1
    estimativas.append(N_hat)

# Preview de algumas estimativas
print([float(x) for x in estimativas[:10]])

```

Explicação:

Foi realizada uma simulação com 10.000 repetições, onde em cada uma foram selecionados aleatoriamente
k=5 tanques distintos de uma população total de
N=100, com numeração de 1 a 100.

Para cada amostra gerada, foi aplicado o estimador clássico:
N^ = M + k / M - 1
onde M é o maior número de série da amostra. Ao final, foi obtida uma lista com 10.000 estimativas diferentes de N^, representando o comportamento do estimador sob repetidas amostragens.

O código mostra as 10 primeiras estimativas, que variam, mas devem ser próximas ao valor real de 100.

###(v) Agora, repita a simulação para diferentes tamanhos de amostra (k = 2, 5, 10, 20). Como o tamanho da amostra influencia a qualidade da estimativa de N?


```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da população
N_real = 100   # Número real de tanques
n_sim = 10000  # Número de simulações
populacao = np.arange(1, N_real + 1)  # População de 1 a 100

# Diferentes tamanhos de amostra
tamanhos_amostra = [2, 5, 10, 20]
estimativas_por_tamanho = {}

for k in tamanhos_amostra:
    estimativas = []
    for _ in range(n_sim):
        amostra = np.random.choice(populacao, size=k, replace=False)
        M = max(amostra)
        N_hat = M + M / k - 1
        estimativas.append(N_hat)

    estimativas_por_tamanho[k] = estimativas

    # Exibindo o histograma para cada tamanho de amostra
    plt.figure(figsize=(12, 5))
    plt.hist(estimativas, bins=50, edgecolor='black', color='lightgray')
    plt.axvline(x=N_real, color='red', linestyle='--', label='Valor real de N = 100')
    plt.title(f"Histograma das estimativas de N̂ (k = {k})")
    plt.xlabel("Valor estimado de N̂")
    plt.ylabel("Frequência")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

    # Média e desvio padrão das estimativas
    media = np.mean(estimativas)
    desvio = np.std(estimativas)
    print(f"Tamanho da amostra k = {k}")
    print(f"Média das estimativas: {media:.2f}")
    print(f"Desvio padrão das estimativas: {desvio:.2f}")
```

Explicação:

O código simula a estimativa de N para os diferentes tamanhos de amostra (k=2,5,10,20) e gera histogramas para cada tamanho de amostra. À medida que o tamanho da amostra aumenta, as estimativas de N se tornam mais precisas, com menos variação e mais concentradas em torno do valor real de N. Isso mostra que amostras maiores fornecem estimativas mais confiáveis e com menor desvio padrão.

# (3) Diferença de médias
### (i) Utilizando o plano amostral descrito acima, produza um histograma que aproxima a distribuição amostral de Md. Dica: repita os passos do plano amostra descrito 10000 vezes. Em seguida, crie o histograma solicitado com os 10000 valores produzidos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros fornecidos
mu1, var1, n1 = 50, 25, 10
mu2, var2, n2 = 40, 24, 8

# Número de simulações
n_sim = 10000

# Simulação
diff_medias = []
for _ in range(n_sim):
    amostra1 = np.random.normal(mu1, np.sqrt(var1), n1)
    amostra2 = np.random.normal(mu2, np.sqrt(var2), n2)
    Md = np.mean(amostra1) - np.mean(amostra2)
    diff_medias.append(Md)

# Histograma
print(Md)
plt.figure(figsize=(12, 5))
plt.hist(diff_medias, bins=50, edgecolor='black', color='lightgray')
plt.title("Distribuição amostral da diferença de médias (Md = M1 - M2)")
plt.xlabel("Diferença de médias")
plt.ylabel("Frequência")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

Explicação:

O histograma mostra a distribuição amostral da diferença de médias (Md = M1 − M2), com a maioria das diferenças concentradas em torno de 10, que é a diferença entre as médias populacionais. A distribuição é aproximadamente normal, com dispersão moderada, indicando que a diferença entre as médias amostrais é precisa e centrada no valor esperado.

##(ii) Usando a aproximação da distribuição amostral de Md obtida no item anterior, compute aproximações para a média e o desvio padrão dessa estatística. Os valores que você obteve são próximos aos fornecidos acima? Explique.

```python
media_diff = np.mean(diff_medias)
desvio_diff = np.std(diff_medias)
print(f"Média simulada de Md: {media_diff:.2f}")
print(f"Desvio padrão simulado de Md: {desvio_diff:.2f}")
```

Explicação:

A média da diferença simulada ficou próxima da diferença teórica de 10, o que confirma que a simulação gerou resultados consistentes com a teoria. Da mesma forma, o desvio padrão simulado ficou muito próximo do valor teórico de 2,35, indicando que a dispersão das médias amostrais está de acordo com o esperado.

Esses resultados mostram que a distribuição amostral da diferença entre as médias segue o comportamento teórico esperado, validando a simulação. A distribuição das diferenças Md se aproxima de uma distribuição normal devido ao Teorema Central do Limite, que se aplica quando o tamanho das amostras é suficientemente grande e as condições de amostragem são atendidas.

```python
import scipy.stats as stats

# Parâmetros da distribuição de Md
mu_md = 10
sigma_md = np.sqrt(5.5)  # = 2.3452

# Valor observado
x = 15

# Calcula o z-score
z = (x - mu_md) / sigma_md

# Calcula a probabilidade P(Md >= 15)
p = 1 - stats.norm.cdf(z)

print(f"z-score: {z:.2f}")
print(f"Probabilidade de Md >= 15: {p:.4f}")
```

Explicação:

Com base na distribuição amostral da diferença de médias Md, que possui média μ Md=10 e desvio padrão σ Md≈2,35, foi calculada a probabilidade de a diferença entre as médias ser 15 ou mais pontos.

Primeiramente, foi calculado o z-score:
z= (15−10) / 2,35 ≈ 2,13

Usando a distribuição normal padrão, a probabilidade correspondente é:
P(M d ≥ 15) ≈ 1,65%

Isso indica que a chance de observar uma diferença de 15 ou mais pontos, caso o medicamento realmente melhore a memória como esperado, é bastante baixa. Portanto, uma diferença tão grande seria considerada um evento raro.

### (iv) Construa um intervalo de confiança para a média de avaliações produzidas por duas populações distintas, P1 e P2, descritas abaixo.
### • P1: todos os recrutadores que ouvem os áudios
### • P2: todos os recrutadores que leem os transcritos
### Note que a resposta a este item devem ser dois intervalos de confiança.
### Nos dois casos, use o nível de confiança de 99%.


```python
import pandas as pd
import numpy as np
import scipy.stats as stats

# Carregar o conjunto de dados
url = "https://raw.githubusercontent.com/AILAB-CEFET-RJ/gcc1625/main/data/SchroederEpley2015data.txt"
df = pd.read_csv(url, sep=",")

# Separar as amostras para P1 (áudio) e P2 (transcrição)
p1 = df[df['CONDITION'] == 1]['Intellect_Rating']  # Recrutadores que ouviram áudio
p2 = df[df['CONDITION'] == 0]['Intellect_Rating']  # Recrutadores que leram transcrição

# Tamanho das amostras
n1 = len(p1)
n2 = len(p2)

# Média e desvio padrão para P1 (áudio)
mean_p1 = np.mean(p1)
std_p1 = np.std(p1, ddof=1)

# Média e desvio padrão para P2 (transcrição)
mean_p2 = np.mean(p2)
std_p2 = np.std(p2, ddof=1)

# Grau de liberdade para a distribuição t (mínimo entre n1-1 e n2-1)
df = min(n1 - 1, n2 - 1)

# Valor crítico t para 99% de confiança (usando distribuição t)
t_99 = stats.t.ppf(0.995, df)

# Erro padrão para cada grupo
se_p1 = std_p1 / np.sqrt(n1)
se_p2 = std_p2 / np.sqrt(n2)

# Intervalos de confiança de 99% para cada grupo
ci_lower_p1 = mean_p1 - t_99 * se_p1
ci_upper_p1 = mean_p1 + t_99 * se_p1

ci_lower_p2 = mean_p2 - t_99 * se_p2
ci_upper_p2 = mean_p2 + t_99 * se_p2

# Exibir os resultados
print(f"Proporção de avaliações para P1 (Áudio): {mean_p1:.4f}")
print(f"Intervalo de confiança de 99% para P1 (Áudio): ({ci_lower_p1:.4f}, {ci_upper_p1:.4f})")

print(f"Proporção de avaliações para P2 (Transcrição): {mean_p2:.4f}")
print(f"Intervalo de confiança de 99% para P2 (Transcrição): ({ci_lower_p2:.4f}, {ci_upper_p2:.4f})")
```

Explicação:

Os resultados mostram que a média das avaliações para o grupo de áudio é 5.63 e para o grupo de transcrição é 3.65. Isso significa que os recrutadores que ouviram o áudio deram avaliações mais altas. O intervalo de confiança de 99% para o grupo áudio é de 4.62 a 6.65 e para o grupo transcrição é de 2.34 a 4.95. Esses intervalos mostram que tem 99% de certeza de que as verdadeiras médias estão dentro dessas faixas. O intervalo para áudio é mais largo, o que indica mais variação nas avaliações. Em resumo, quem ouviu o áudio avaliou mais alto, mas com mais variação, enquanto quem leu a transcrição deu notas mais consistentes, mas mais baixas.

##(ii) Produza o intervalo de confiança solicitado, usando o nível de confiança 90%. Junto com o resultado, forneça também uma análise.

```python
import numpy as np
import scipy.stats as stats

# Proporção de canhotos
total_students = len(df)
left_handed_count = len(left_handed)
p_hat = left_handed_count / total_students

# Tamanho da amostra
n = total_students

# Z para nível de confiança de 90%
z_90 = stats.norm.ppf(1 - 0.05)  # 90% de confiança, Z = 1.645

# Erro padrão para proporção
se = np.sqrt((p_hat * (1 - p_hat)) / n)

# Intervalo de confiança de 90%
ci_lower = p_hat - z_90 * se
ci_upper = p_hat + z_90 * se

# Exibindo o intervalo de confiança
print(f"Proporção de canhotos: {p_hat:.4f}")
print(f"Intervalo de confiança de 90% para a proporção de canhotos: ({ci_lower:.4f}, {ci_upper:.4f})")

```

O valor crítico da distribuição normal para um intervalo de confiança de 90% é aproximadamente 1.645, e isso é usado para calcular o intervalo de confiança, aplicando a fórmula
